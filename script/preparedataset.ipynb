```python
import pandas as pd
import numpy as np
import warnings
from statistics import mean


warnings.filterwarnings("ignore")
pd.set_option('display.max_columns', None)

datadir = "../UFC-Web-Scraping/ufc_web_scraper/scraped_files/"

ufc_fighter_stat_df  = pd.read_csv(f"{datadir}/ufc_fight_stat_data.csv")
fight_data_df = pd.read_csv(f"{datadir}/ufc_fight_data.csv")
ufc_fighters =  pd.read_csv(f"{datadir}/ufc_fighter_data.csv")
event_data_df =  pd.read_csv(f"{datadir}/ufc_event_data.csv") 
rankings_history =  pd.read_csv(f"rankings_history.csv") 
```

```python
ufc_fighter_stat_df = ufc_fighter_stat_df.rename(columns={'fighter_id': 'fighter_fullname'})
```

```python
main_df = pd.merge(ufc_fighter_stat_df,ufc_fighters,how='left',left_on='fighter_detail_url', right_on='fighter_url') 
main_df = pd.merge(main_df,fight_data_df,how='left',on='fight_url') 
main_df = pd.merge(main_df,event_data_df,how='left',left_on='eventurl',right_on='event_url') 
```

```python
# Add age
main_df['fighter_dob'] = pd.to_datetime(main_df['fighter_dob'])
main_df['event_date'] = pd.to_datetime(main_df['event_date'])
main_df['age_on_attemp']  = main_df['event_date'] - main_df['fighter_dob']
main_df['age_on_attemp'] = main_df['age_on_attemp'].dt.days / 365
main_df['fighter_stance'] =main_df['fighter_stance'].fillna("Other")
# main_df = main_df[main_df['age_on_attemp'].notna()]
main_df['age_on_attemp'] = main_df['age_on_attemp'].round(2) 
main_df['temp_count'] = 1

#  cumulative win

def wincount(x):
    # print(x)
    if x['winner'] == x['fighter_fullname']:
        return 1
    else:
        return 0

# winloss dedct
def winlossstatscore(x):
    # print(x)
    if x['winner'] == x['fighter_fullname']:
        return 1
    else:
        return -1
main_df['win_count']  = main_df[['winner','fighter_fullname']].apply(wincount, axis=1)
main_df['win_loss_stat_score']  = main_df[['winner','fighter_fullname']].apply(winlossstatscore, axis=1)
main_df = main_df.reset_index()
main_df = main_df.sort_values(by=['event_date','index'], ascending=[True, False])
main_df = main_df.set_index("index")
main_df = main_df.reset_index(drop = True)
main_df['win_acccumulate'] = main_df[['fighter_fullname', 'win_count']].groupby('fighter_fullname').cumsum()
main_df['match_count_acccumulate'] = main_df[['fighter_fullname', 'temp_count']].groupby('fighter_fullname').cumsum()
main_df['win_loss_stat_score_acccumulate'] = main_df[['fighter_fullname', 'win_loss_stat_score']].groupby('fighter_fullname').cumsum()
```

```python
main_df['win_ratio'] = main_df['win_acccumulate'] / main_df['match_count_acccumulate']
```

```python
#  Match ranking on attemp

rank_list = []
for i in tqdm(range(len(main_df))):
    ranking_check = None
    temp_df = main_df[i:i+1]
    temp_df =  pd.merge(temp_df,rankings_history,how='left',left_on=['fighter_fullname','weight_class'],right_on=['fighter','weightclass'])
    if str(temp_df['date'][0:1].values[0]) == "nan":
        rank_list.append(ranking_check)
        continue
    temp_df['date'] = pd.to_datetime(temp_df['date'])
    temp_df['date_diff']  = temp_df['date'] - temp_df['event_date']
    temp_df['date_diff'] = temp_df['date_diff'].dt.days
    temp_df = temp_df[temp_df['date_diff'] <= 0]
    if len(temp_df) != 0 :
        ranking_check = temp_df[-1:]['rank'].values[0]
        rank_list.append(ranking_check)
    else:
        rank_list.append(None)

main_df['ranking_on_attemp'] = rank_list
```

```python
def prepare_feature(fight_url,main_df):

    # previous_stat
    temp_df = main_df[main_df['fight_url'] == fight_url]

    fdict = {}
    f1name =  temp_df['f_1'].values[0]
    f2name =  temp_df['f_2'].values[0]
    fdict[f"fight_round"] = temp_df['num_rounds'].values[0]
    fdict[f"event_city"] = temp_df['event_city'].values[0]
    fdict[f"event_state"] = temp_df['event_state'].values[0]
    fdict[f"event_date"] = temp_df['event_date'].values[0]
    fdict[f"winner"] = temp_df['winner'].values[0]
    
    for ind, row in temp_df.iterrows():
        # print(ind)
    
        if row['fighter_fullname'] == f1name:
            pref = "f_1"
            
    
        elif row['fighter_fullname'] == f2name:
            pref = "f_2"

        fdict[f"{pref}_fighter_fullname"] = row['fighter_fullname']
        fdict[f"{pref}_age"] = row['age_on_attemp']
        fdict[f"{pref}_height"] = row['fighter_height_cm']
        fdict[f"{pref}_weight"] = row['fighter_weight_lbs']
    
        fdict[f"{pref}_fighter_stance"] = row['fighter_stance']
        fdict[f"{pref}_weight_class"] = row['weight_class']
        fdict[f"{pref}_gender"] = row['gender']
        
        #  feature  
        fdict[f"{pref}_prev_win_streak_continuous"] = 0
        fdict[f"{pref}_prev_loss_streak_continuous"] = 0
        fdict[f"{pref}_total_match"] = 0
        fdict[f"{pref}_win_acccumulate"] = 0
        fdict[f"{pref}_win_ratio"] = 0
        fdict[f"{pref}_prev_result"] = None
        fdict[f"{pref}_prev_result_details"] = None
        fdict[f"{pref}_prev_finish_round"] = None
        fdict[f"{pref}_prev_finish_time"] = None
        fdict[f"{pref}_prev_weight_class"] = None
        fdict[f"{pref}_avg_total_strike_succ_ratio"] = 0
        fdict[f"{pref}_avg_sig_strikes_succ_ratio"] = 0
        fdict[f"{pref}_avg_takedown_succ_ratio"] = 0
        fdict[f"{pref}_avg_knockdown_ratio"] = 0
        fdict[f"{pref}_avg_submission_ratio"] = 0
        fdict[f"{pref}_avg_reversal_ratio"] = 0
        fdict[f"{pref}_avg_head_strikes_succ_ratio"] = 0
        fdict[f"{pref}_avg_body_strikes_succ_ratio"] = 0
        fdict[f"{pref}_avg_leg_strikes_succ_ratio"] = 0
        fdict[f"{pref}_avg_distance_strikes_succ_ratio"] = 0
        fdict[f"{pref}_avg_clinch_strikes_succ_ratio"] = 0
        fdict[f"{pref}_avg_ground_strikes_succ_ratio"] = 0
    
        prev_df = main_df[(main_df.index < ind) & (main_df['fighter_fullname'] == row['fighter_fullname'])]
        # print(prev_df)
        #  Check win streak
        prev_win_streak_continuous = 0
        prev_loss_streak_continuous = 0
        for k in prev_df['win_count'].tolist():
            if int(k) == 1:
                prev_win_streak_continuous += 1
            else:
                prev_win_streak_continuous = 0
    
        for k in prev_df['win_count'].tolist():
            if int(k) == 0:
                prev_loss_streak_continuous += 1
            else:
                prev_loss_streak_continuous = 0
    
        fdict[f"{pref}_prev_win_streak_continuous"] = prev_win_streak_continuous
        fdict[f"{pref}_prev_loss_streak_continuous"] = prev_loss_streak_continuous
    
        if len(prev_df) > 0 :
    
            fdict[f"{pref}_prev_result"] = prev_df.tail(1)['result'].values[0]
            fdict[f"{pref}_prev_result_details"] = prev_df.tail(1)['result_details'].values[0]
            fdict[f"{pref}_prev_finish_round"] = prev_df.tail(1)['finish_round'].values[0]
            fdict[f"{pref}_prev_finish_time"] = prev_df.tail(1)['finish_time'].values[0]
            fdict[f"{pref}_total_match"] = prev_df.tail(1)['match_count_acccumulate'].values[0]
            fdict[f"{pref}_win_acccumulate"] = prev_df.tail(1)['win_acccumulate'].values[0]
            fdict[f"{pref}_win_ratio"] = prev_df.tail(1)['win_ratio'].values[0]
    
            prev_df['avg_total_strike_succ_ratio'] = prev_df['total_strikes_succ'] / prev_df['total_strikes_att']
            prev_df['avg_sig_strikes_succ_ratio'] = prev_df['sig_strikes_succ'] / prev_df['sig_strikes_att']
            prev_df['avg_takedown_ratio'] = prev_df['takedown_succ'] / prev_df['takedown_att']
            prev_df['avg_head_strikes_succ_ratio'] = prev_df['head_strikes_succ'] / prev_df['head_strikes_att']
            prev_df['avg_body_strikes_succ_ratio'] = prev_df['body_strikes_succ'] / prev_df['body_strikes_att']
            prev_df['avg_leg_strikes_succ_ratio'] = prev_df['leg_strikes_succ'] / prev_df['leg_strikes_att']
            prev_df['avg_distance_strikes_succ_ratio'] = prev_df['distance_strikes_succ'] / prev_df['distance_strikes_att']
            prev_df['avg_clinch_strikes_succ_ratio'] = prev_df['clinch_strikes_succ'] / prev_df['clinch_strikes_att']
            prev_df['avg_ground_strikes_succ_ratio'] = prev_df['ground_strikes_succ'] / prev_df['ground_strikes_att']
    
            try:
                fdict[f"{pref}_avg_total_strike_succ_ratio"] = mean(prev_df[prev_df['avg_total_strike_succ_ratio'].notna()]['avg_total_strike_succ_ratio'].tolist())
            except:
                pass
            try:
                fdict[f"{pref}_avg_sig_strikes_succ_ratio"] = mean(prev_df[prev_df['avg_sig_strikes_succ_ratio'].notna()]['avg_sig_strikes_succ_ratio'].tolist())
            except:
                pass
            try:
                fdict[f"{pref}_avg_takedown_succ_ratio"] = mean(prev_df[prev_df['avg_takedown_ratio'].notna()]['avg_takedown_ratio'].tolist())
            except:
                pass
            try:
                fdict[f"{pref}_avg_knockdown_ratio"] = mean(prev_df[prev_df['knockdowns'].notna()]['knockdowns'].tolist())
            except:
                pass
            try:
                fdict[f"{pref}_avg_submission_ratio"] = mean(prev_df[prev_df['submission_att'].notna()]['submission_att'].tolist())
            except:
                pass
            try:
                fdict[f"{pref}_avg_reversal_ratio"] = mean(prev_df[prev_df['reversals'].notna()]['reversals'].tolist())
            except:
                pass
            try:
                fdict[f"{pref}_avg_head_strikes_succ_ratio"] = mean(prev_df[prev_df['avg_head_strikes_succ_ratio'].notna()]['avg_head_strikes_succ_ratio'].tolist())
            except:
                pass
            try:
                fdict[f"{pref}_avg_body_strikes_succ_ratio"] = mean(prev_df[prev_df['avg_body_strikes_succ_ratio'].notna()]['avg_body_strikes_succ_ratio'].tolist())
            except:
                pass
            try:
                fdict[f"{pref}_avg_leg_strikes_succ_ratio"] = mean(prev_df[prev_df['avg_leg_strikes_succ_ratio'].notna()]['avg_leg_strikes_succ_ratio'].tolist())
            except:
                pass
            try:
                fdict[f"{pref}_avg_distance_strikes_succ_ratio"] = mean(prev_df[prev_df['avg_distance_strikes_succ_ratio'].notna()]['avg_distance_strikes_succ_ratio'].tolist())
            except:
                pass
            try:
                fdict[f"{pref}_avg_clinch_strikes_succ_ratio"] = mean(prev_df[prev_df['avg_clinch_strikes_succ_ratio'].notna()]['avg_clinch_strikes_succ_ratio'].tolist())
            except:
                pass
            try:
                fdict[f"{pref}_avg_ground_strikes_succ_ratio"] = mean(prev_df[prev_df['avg_ground_strikes_succ_ratio'].notna()]['avg_ground_strikes_succ_ratio'].tolist())
            except:
                pass
            try:
                fdict[f"{pref}_avg_ground_strikes_succ_ratio"] = mean(prev_df[prev_df['avg_ground_strikes_succ_ratio'].notna()]['avg_ground_strikes_succ_ratio'].tolist())
            except:
                pass


    
    return fdict
```

```python
#  Prediction WinLoss
adata = []
all_unique_fight_url = main_df['fight_url'].unique().tolist()
for i in tqdm(all_unique_fight_url):
    adata.append(prepare_feature(i,main_df))
```

```python
dfItem = pd.DataFrame.from_records(adata)

dfItem = dfItem[(dfItem['f_2_age'].notna()) & (dfItem['f_1_age'].notna()) & (dfItem['f_2_weight_class'].notna()) & (dfItem['f_1_weight_class'].notna()) ]

dfItem = dfItem[dfItem['fight_round'].isin(["3","5"])]
```

```python
odd_data =  pd.read_csv(f"{datadir}/ufc-master.csv") 
odd_data = odd_data[['RedFighter','BlueFighter','RedOdds','BlueOdds','Date','Location']]
odd_data['Location'] = odd_data['Location'].str.split(",").str[0]
dfItem['event_date_str'] = dfItem['event_date'].astype(str)
dfItem = pd.merge(dfItem,odd_data,how='left',left_on=['f_1_fighter_fullname','f_2_fighter_fullname','event_date_str','event_city'], right_on=['RedFighter','BlueFighter','Date','Location']) 
```

```python
#  Prepare ML Dataset

dataset_df = pd.DataFrame()
for ind, row in dfItem.iterrows():
    ttdf = pd.DataFrame()
    label = 0
    if row['f_1_fighter_fullname'] == row['winner'] :
        label = 1
    ttdf['id'] = [ind]
    ttdf['f_1_odd'] = [row['RedOdds']]
    ttdf['f_2_odd'] = [row['BlueOdds']]
    ttdf['fight_round'] = [row['fight_round']]
    ttdf['event_city'] = [row['event_city']]
    ttdf['event_state'] = [row['event_state']]
    ttdf['f_1_gender'] = [row['f_1_gender']]
    ttdf['f_2_gender'] = [row['f_2_gender']]
    ttdf['f_1_weight_class'] = [row['f_1_weight_class']]
    ttdf['f_2_weight_class'] = [row['f_2_weight_class']]
    ttdf['f_1_fighter_stance'] = [row['f_1_fighter_stance']]
    ttdf['f_2_fighter_stance'] = [row['f_2_fighter_stance']]
    ttdf['f_1_prev_result'] = [row['f_1_prev_result']]
    ttdf['f_2_prev_result'] = [row['f_2_prev_result']]
    
    ttdf['age_diff'] = [row['f_1_age'] - row['f_2_age']]
    ttdf['height_diff'] = [row['f_1_height'] - row['f_2_height']]
    ttdf['weight_diff'] = [row['f_1_weight'] - row['f_2_weight']]
    ttdf['win_ratio_diff'] = [row['f_1_win_ratio'] - row['f_2_win_ratio']]
    ttdf['prev_win_streak_continuous_diff'] = [row['f_1_prev_win_streak_continuous'] - row['f_2_prev_win_streak_continuous']]
    ttdf['prev_loss_streak_continuous_diff'] = [row['f_1_prev_loss_streak_continuous'] - row['f_2_prev_win_streak_continuous']]
    ttdf['avg_total_strike_succ_ratio_diff'] = [row['f_1_avg_total_strike_succ_ratio'] - row['f_2_avg_total_strike_succ_ratio']]
    ttdf['avg_sig_strikes_succ_ratio_diff'] = [row['f_1_avg_sig_strikes_succ_ratio'] - row['f_2_avg_sig_strikes_succ_ratio']]
    ttdf['avg_submission_ratio_diff'] = [row['f_1_avg_submission_ratio'] - row['f_2_avg_submission_ratio']]
    ttdf['avg_reversal_ratio_diff'] = [row['f_1_avg_reversal_ratio'] - row['f_2_avg_reversal_ratio']]
    ttdf['avg_head_strikes_succ_ratio_diff'] = [row['f_1_avg_head_strikes_succ_ratio'] - row['f_2_avg_head_strikes_succ_ratio']]
    ttdf['avg_body_strikes_succ_ratio_diff'] = [row['f_1_avg_body_strikes_succ_ratio'] - row['f_2_avg_body_strikes_succ_ratio']]
    ttdf['avg_leg_strikes_succ_ratio_diff'] = [row['f_1_avg_leg_strikes_succ_ratio'] - row['f_2_avg_leg_strikes_succ_ratio']]
    ttdf['avg_distance_strikes_succ_ratio_diff'] = [row['f_1_avg_distance_strikes_succ_ratio'] - row['f_2_avg_distance_strikes_succ_ratio']]
    ttdf['avg_clinch_strikes_succ_ratio_diff'] = [row['f_1_avg_clinch_strikes_succ_ratio'] - row['f_2_avg_clinch_strikes_succ_ratio']]
    ttdf['avg_ground_strikes_succ_ratio_diff'] = [row['f_1_avg_ground_strikes_succ_ratio'] - row['f_2_avg_ground_strikes_succ_ratio']]
    ttdf['label'] = [label]
    dataset_df = pd.concat([dataset_df,ttdf])
```
